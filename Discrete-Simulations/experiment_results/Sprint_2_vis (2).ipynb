```python
# df = pd.read_csv('test_experiments_more_levels.csv')
# df = pd.read_csv('test_experiments_all_3.csv') 
path = '/content/drive/MyDrive/colabdata/test_experiments.csv'
mc = '/content/drive/MyDrive/colabdata/test_experiments_monte_carlo.csv'
df_mc
df = pd.read_csv(path)
```

```python
df_mc = pd.read_csv(mc)
df_mc
```

```python
df2 = df_mc.groupby(['steps_per_episode', 'discount_factor'], as_index=False).mean()
df2
```

```python
fig = make_subplots(rows=1, cols=1)


# Subplot 1
fig.add_trace(
    go.Bar(
        name="Cleaned",
        x=[ df2.steps_per_episode,df2.discount_factor],
        y=df2["cleaned"], marker_color='crimson'
    ),
    row=1,
    col=1
)


fig.add_trace(
    go.Bar(
        name="Efficiency",
        x=[ df2.steps_per_episode,df2.discount_factor],
        y=df2["efficiency"],marker_color='blue'
    ),
    row=1,
    col=1,
)

fig.update_layout(title={
        'text': "Cleaned vs efficiency percentages of monte carlo robot",
        'y':0.9,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'}, 
        autosize=True,
        width=1000,
        height=400,
        font = dict(size = 12))

# fig.update_xaxes(tickangle=30)

fig.show()
```
